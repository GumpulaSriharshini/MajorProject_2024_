```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load the tokenizer and model
model_name = "facebook/nllb-200-distilled-600M"  # Smaller model for efficiency
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Define the input text and languages
input_text = "Hello, how are you?"  # Text to translate
source_lang = "eng_Latn"  # Source language code
target_lang = "hin_Deva"  # Target language code (Hindi in Devanagari script)

# Tokenize the input text
inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)

# Manually set the forced BOS token for the target language
if target_lang in tokenizer.additional_special_tokens:
    forced_bos_token_id = tokenizer.convert_tokens_to_ids(target_lang)
else:
    raise ValueError(f"Target language '{target_lang}' is not supported by the tokenizer.")

        # Generate the translation
outputs = model.generate(
    inputs["input_ids"],
    forced_bos_token_id=forced_bos_token_id,
    max_length=100,
    num_beams=5
 )

# Decode the translated text
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"Translated Text: {translated_text}")
```
